```python
nf = df.to_numpy()
```

```python
len(nf)
```

```python
nf[0]
```

```python
new_nf = []
new_df_columns = ['ch', 'pyn', 'py', 'en']
```

```python
import jellyfish

for idx, word in enumerate(nf[:, 0]):
    pinyin_translation = pc.translate_word(word)
    if pinyin_translation != None:
        mts_part = nf[idx, 2].replace("/", " / ")
        pinyin_part = " / ".join(pinyin_translation)
        
        if jellyfish.jaro_distance(mts_part, pinyin_part) >= 0.8:
            if len(pinyin_part) > len(mts_part):
                english_translation = pinyin_part
            else:
                english_translation = mts_part
        else:
            english_translation = mts_part + "\n" + pinyin_part
        
        row = [word, nf[idx, 1].replace("/", " / "), pinyin.get(word), english_translation]
        new_nf.append(row)
    else:
        print("none")
```

```python
new_df = pd.DataFrame(new_nf, columns=new_df_columns)
```
